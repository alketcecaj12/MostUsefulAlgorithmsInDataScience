```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn_extra.cluster import KMedoids
from sklearn.datasets import make_blobs
```

```python
data = pd.DataFrame({"Age":[   20,   19,   23,  24,  60,     59,  57,   54,    89,    90,     92,      94],
                     "Money":[2000, 1900, 2300, 2600, 60000, 59000, 59000, 54000, 890000, 900000, 920000, 90000000],
                     "House": [20,   21,    22,   23,  70,    75,    58, 69,     150,   200,   180, 60000]})
```

```python
X = data.values
X
```

```python
cobj = KMedoids(n_clusters=3).fit(X)
labels = cobj.labels_
```

```python
labels
```

```python
data["clusters"] = labels
data
```

```python
# importing mplot3d toolkits
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
 
fig = plt.figure()
 
# syntax for 3-D projection
ax = plt.axes(projection ='3d')
 
# defining axes
z = data[data["clusters"]==0].T.values[0:3,0:3]
x = data[data["clusters"]==2].T.values[0:3,0:3]
y = data[data["clusters"]==1].T.values[0:3,0:3]
x
c = x + y
ax.scatter(x, y, z, c = c)
 
# syntax for plotting
ax.set_title('3d Scatter plot geeks for geeks')
plt.show()
```
